```python
import matplotlib.pyplot as plt
```

```python
x=['Aleena','Aneeta','Deepthi','Aneeka']
y=[20,36,12,40]
y1=[165,130,140,160]
plt.plot(x,y)
plt.plot(x,y1)
plt.xlabel('name')
plt.ylabel('age')
```

```python
x=['Aleena','Aneeta','Deepthi','Aneeka']
y=[20,36,12,40]
y1=[165,130,140,160]
plt.plot(x,y,linestyle='--',marker='p',c='red')
plt.plot(x,y1)
plt.xlabel('name')
plt.ylabel('age')
```

```python
x=['Aleena','Aneeta','Deepthi','Aneeka']
y=[20,36,12,40]
y1=[165,130,140,160]
plt.subplot(1,2,1)
plt.plot(x,y,linestyle='--',marker='p',c='red')
plt.subplot(1,2,2)
plt.plot(x,y1)
plt.xlabel('name')
plt.ylabel('age')
```

```python
e=np.array([2,4,6,8,2])
names=np.array(['A','B','C','D','E'])
plt.pie(e,labels=names,startangle=90)
```

```python
e=np.array([2,4,6,8,2])
names=np.array(['A','B','C','D','E'])
explode=[.2,0,0,0,0]
colors=np.array(['red','green','yellow','violet','orange'])

plt.pie(e,labels=names,startangle=90,explode=explode,shadow=True,colors=colors)
plt.legend(title='Fruits')
```
